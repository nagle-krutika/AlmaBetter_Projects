## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1: Seat Comfort and Referral

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

- H0: There is no significant difference in seat comfort ratings between passengers who recommend the airline and those who do not.
- H1: There is a significant difference in seat comfort ratings between passengers who recommend the airline and those who do not.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

from scipy import stats

# Create two groups
recommenders = df[df['recommended'] == 1]['seat_comfort'].dropna()
non_recommenders = df[df['recommended'] == 0]['seat_comfort'].dropna()

# Perform t-test (assuming unequal variances)
t_stat, p_value = stats.ttest_ind(recommenders, non_recommenders, equal_var=False)

print("Hypothesis 1: Seat Comfort and Referral")
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}\n")

# Interpret results
alpha = 0.05
if p_value < alpha:
    print("Reject H0: There is a significant difference in seat comfort ratings.")
else:
    print("Fail to reject H0: No significant difference in seat comfort ratings.\n")
````

##### Which statistical test have you done to obtain P-Value?

Independent Samples t-test

##### Why did you choose the specific statistical test?

- Comparing means of two independent groups (recommenders vs. non-recommenders)
- 'seat_comfort' is a numerical variable
- Assuming normal distribution of seat comfort ratings
- Groups are independent (a passenger is either a recommender or not)

### Hypothetical Statement - 2: Traveller Type and Entertainment Rating

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

- H0: There is no association between traveller type (business vs. leisure) and their rating of in-flight entertainment.
- H1: There is an association between traveller type (business vs. leisure) and their rating of in-flight entertainment.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

# Bin entertainment ratings into categories
df['entertainment_category'] = pd.qcut(df['entertainment'], q=3, labels=['Low', 'Medium', 'High'])

# Create a contingency table
contingency = pd.crosstab(df['traveller_type'], df['entertainment_category'])

# Perform Chi-Square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency)

print("Hypothesis 2: Traveller Type and Entertainment Rating")
print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")
print(f"Degrees of freedom: {dof}\n")

# Interpret results
if p_value < alpha:
    print("Reject H0: There is an association between traveller type and entertainment rating.")
else:
    print("Fail to reject H0: No association between traveller type and entertainment rating.\n")
````

##### Which statistical test have you done to obtain P-Value?

Chi-Square Test for Independence

##### Why did you choose the specific statistical test?

- Testing association between two categorical variables
- 'traveller_type' is categorical (business, leisure)
- 'entertainment' rating can be treated as categorical (e.g., binned into "High", "Medium", "Low")
- Large sample size (> 1000) for reliable chi-square approximation

### Hypothetical Statement - 3: Overall Rating and Value for Money



#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

- H0: There is no correlation between a passenger's overall rating of the airline and their perception of value for money.
- H1: There is a correlation between a passenger's overall rating of the airline and their perception of value for money.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

# Drop any rows with missing values
df_clean = df[['overall', 'value_for_money']].dropna()

# Calculate Spearman's rank correlation
correlation, p_value = stats.spearmanr(df_clean['overall'], df_clean['value_for_money'])

print("Hypothesis 3: Overall Rating and Value for Money")
print(f"Spearman's rank correlation: {correlation}")
print(f"P-value: {p_value}\n")

# Interpret results
if p_value < alpha:
    print("Reject H0: There is a correlation between overall rating and value for money.")
else:
    print("Fail to reject H0: No correlation between overall rating and value for money.")
````

##### Which statistical test have you done to obtain P-Value?

Spearman's Rank Correlation Test

##### Why did you choose the specific statistical test?

- Testing correlation between two ordinal variables
- Both 'overall' rating and 'value_for_money' are ordinal (ranked on a scale)
- No assumption of normal distribution needed
- Can handle non-linear relationships

# **Pipeline**

````python

# Creating a Pipeline

from joblib import Memory

cachedir = '.cache'
memory = Memory(location=cachedir, verbose=0)

pipelines = {
    'lr' : Pipeline([
        ('scaler', StandardScaler()),
        ('model', LogisticRegression())
    ]),
    'rfc' : Pipeline([
        ('scaler', StandardScaler()),
        ('model', RandomForestClassifier())
    ]),
    'dtc' : Pipeline([
        ('scaler', StandardScaler()),
        ('model', DecisionTreeClassifier())
    ]),
    'svm' : Pipeline([
        ('scaler', StandardScaler()),
        ('model', SVC())
    ]),
    'knn' : Pipeline([
        ('scaler', StandardScaler()),
        ('model', KNeighborsClassifier())
    ])
}

# Define parameter grid for each algorithm
param_grids = {
    'lr': [
        {
            'scaler': [MinMaxScaler(), StandardScaler()]
        }
    ],
    'rfc': [
        {
            'scaler': [MinMaxScaler(), StandardScaler()]

        }
    ],
    'dtc': [
        {
            'scaler': [MinMaxScaler(), StandardScaler()]
        }
    ],
    'svm': [
        {
            'scaler': [MinMaxScaler(), StandardScaler()]

        }
    ],
    'knn': [
        {
            'scaler': [MinMaxScaler(), StandardScaler()],

        }
    ]
}

best_models = {}

# Run the Pipeline
for algo in pipelines.keys():
    print("*"*10, algo, "*"*10)
    random_search = RandomizedSearchCV(estimator=pipelines[algo],
                               param_distributions=param_grids[algo],
                               cv=5,
                               scoring='accuracy',
                               return_train_score=True,
                               verbose=1
                              )

    %time random_search.fit(x_train, y_train)

    print('Train Score: ', random_search.best_score_)
    print('Test Score: ', random_search.score(x_test, y_test))
    print('Best_Params:', random_search.best_params_)

    best_models[algo] = random_search.best_estimator_
    print()
````
