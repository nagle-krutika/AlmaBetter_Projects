## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# Code to copy data from main dataset
dataset=df.copy()
dataset.head()
````

````python
# Code to fill null values of column 'Rating'

dataset.Rating.fillna(dataset.Rating.mean(),inplace=True)
dataset.Rating.isnull().sum()
````

````python
# Code to remove '0' value row form 'Type' column

print(dataset.Type.value_counts())

dataset=dataset[dataset["Type"] != "0"]
dataset.Type.value_counts()
````

````python
# Code to convert object type to float type in 'reviews' column

print(dataset.Reviews.info())
l1=[]

for i in dataset['Reviews']:
    if "M" in i:
        i=i.replace("M","")
        l1.append(float(i))
    if type(i) == str:
        l1.append(float(i))

dataset['Reviews']=l1
````

````python
# Data after conversion

dataset.Reviews.info()
````

````python
#Code to replace object values to float values

dataset.Size.info()
l2=[]
for i in dataset['Size']:
    if "M" in str(i):
        i=i.replace("M","")
    if "k" in str(i):
        i=i.replace("k","")
    if "+" in str(i):
        i=i.replace("+","")
    if "," in str(i):
        i=i.replace(",","")
    if type(i) == str:
        i=float(i)
        l2.append(i)
    else:
      l2.append(i)
dataset['Size']=l2
````

````python
# Replacing 'Varies with device' with '0'
dataset['Size'].replace("Varies with device","0",inplace = True)
````

````python
dataset.Size.info()
````

````python
#Code to replace object values to int values

dataset.Installs.info()

l1=[]

for i in dataset['Installs']:
    if "+" in str(i):
        i=i.replace("+","")
    if "," in str(i):
        i=i.replace(",","")
    if i=="Free":
        i=0
    if type(i) == str:
        i=int(i)
        l1.append(i)
    else:
      l1.append(i)

dataset['Installs']=l1
````

````python
dataset.Installs.info()
````

````python
# Code to replace object values to float values

dataset.Price.info()

l1=[]

for i in dataset['Price']:
    if "$" in i:
        i=i.replace("$","")
    if i == "Everyone":
        i=0
    if type(i) == str:
        i=float(i)
    l1.append(i)

dataset['Price']=l1
````

````python
dataset.Price.info()
````

##**Label Encoding**

````python
# Code to do label encoding in type column

dataset.Type.replace({"Free":0, "Paid":1},inplace=True)
dataset.Type.value_counts()
````

````python
dataset.head(3)
````

````python
dataset.shape
````

````python
# Code to remove duplicate values

dataset=dataset.drop_duplicates()
dataset.shape
````

````python
# Code to remove null/missing values

dataset=dataset.dropna()
dataset.shape
````

````python
dataset.isnull().sum()
````
